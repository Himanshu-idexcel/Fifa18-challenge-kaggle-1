```python
top_cb=fifa_18.groupby('Name')['CB'].mean().sort_values(ascending=False).iloc[: 10].index.tolist()
top_rb=fifa_18.groupby('Name')['RB'].mean().sort_values(ascending=False).iloc[: 10].index.tolist()
top_lb=fifa_18.groupby('Name')['LB'].mean().sort_values(ascending=False).iloc[: 10].index.tolist()
top_lm=fifa_18.groupby('Name')['LM'].mean().sort_values(ascending=False).iloc[: 10].index.tolist()
top_rm=fifa_18.groupby('Name')['RM'].mean().sort_values(ascending=False).iloc[: 10].index.tolist()
top_cdm=fifa_18.groupby('Name')['CDM'].mean().sort_values(ascending=False).iloc[: 10].index.tolist()
top_st=fifa_18.groupby('Name')['ST'].mean().sort_values(ascending=False).iloc[: 10].index.tolist()
top_rw=fifa_18.groupby('Name')['RW'].mean().sort_values(ascending=False).iloc[: 10].index.tolist()
top_lw=fifa_18.groupby('Name')['LW'].mean().sort_values(ascending=False).iloc[: 10].index.tolist()
```

```python
best_players = pd.DataFrame({'CB': top_cb,'RB':top_rb,'LB':top_lb,'LM':top_lm,'RM':top_rm,'CDM':top_cdm,'St':top_st,'RW':top_rw,'LW':top_lw})
```

```python
#here is the list for top players according to their field positions.
best_players
```
